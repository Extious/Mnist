### 二、MNIST-GAN网络构建

GAN模型包含两部分

- 生成模型（G，Generative Model）
- 判别模型（D，Discriminative Model）

在二者的互相博弈学习中产生预期的输出。原始 GAN 理论中，并不要求 G 和 D 都是神经网络，只需要是能拟合相应生成和判别的函数即可。但实用中一般均使用深度神经网络作为 G 和 D 。

##### 生成模型

生成器的目标是生成尽可能逼真的数据来欺骗判别器。它接收一个随机噪声向量作为输入，并通过一系列的神经网络层将其转换为数据样本（例如图像）。生成器试图学习真实数据的分布，以便生成的数据样本与真实数据无法区分。

生成器采用多层感知机，输入一个1x100的随机噪声，经过4个全连接层，采用Leaky Relu激活函数，最后生成一个1x28x28的数字图像
生成器的网络结构如下：

1. **输入层（Input Layer）:**
   输入大小为input_size，这意味着生成器接收一个大小为input_size=100的随机噪声向量作为输入。
2. **隐藏层（Hidden Layers）:**
   第一隐藏层（fc1）: 一个全连接层（nn.Linear），将输入向量从input_size=100维映射到128维。激活函数为Leaky ReLU，负斜率参数为0.2。
   第二隐藏层（fc2）: 另一个全连接层，将前一层的输出从128维映射到256维（128 * 2）。同样使用Leaky ReLU激活函数。
   第三隐藏层（fc3）: 第三个全连接层，将前一层的输出从256维映射到512维（256 * 2）。同样使用Leaky ReLU激活函数。
3. **输出层（Output Layer）:**
   最后一层（fc4）: 一个全连接层，将前一层的输出从512维映射到output_size维。激活函数为Tanh，将输出值限制在[-1, 1]之间。



```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchsummary


class Generator(nn.Module):
    def __init__(self, input_size, output_size):
        super(Generator,self).__init__()
        self.fc1 = nn.Linear(input_size,128)
        self.fc2 = nn.Linear(self.fc1.out_features, self.fc1.out_features * 2)
        self.fc3 = nn.Linear(self.fc2.out_features, self.fc2.out_features * 2)
        self.fc4 = nn.Linear(self.fc3.out_features, output_size)
    def forward(self,x):
        x = F.leaky_relu(self.fc1(x),0.2)
        x = F.leaky_relu(self.fc2(x), 0.2)
        x = F.leaky_relu(self.fc3(x), 0.2)
        x = torch.tanh(self.fc4(x))
G = Generator(input_size=100, output_size=28 * 28)
print("summary of G:")
torchsummary.summary(G,input_size=[(1,100)],batch_size=32,device="cpu")
```

##### 判别模型

判别器的目标是区分输入的数据是真实的（来自真实数据集）还是生成的（来自生成器）。它接收数据样本作为输入，并输出一个概率值，表示该样本是真实数据的概率。判别器通过学习真实数据和生成数据的特征来提高其判别能力。

判别器采用多层感知机，输入一个1x784的MNIST数字图像，经过4个全连接层，采用Leaky Relu激活函数，最后输出属于MNIST数据集的概率。

判别器的网络结构如下：

1. **输入层（Input Layer）:**
   输入大小为input_size，这意味着判别器接收一个大小为input_size的数据样本作为输入。这个样本可以是来自真实数据集的图像，也可以是生成器生成的图像。
2. **隐藏层（Hidden Layers）:**
   第一隐藏层（fc1）: 一个全连接层（nn.Linear），将输入向量从input_size维映射到512维。激活函数为Leaky ReLU，负斜率参数为0.2。
   第二隐藏层（fc2）: 另一个全连接层，将前一层的输出从512维映射到256维（512 // 2）。同样使用Leaky ReLU激活函数。
   第三隐藏层（fc3）: 第三个全连接层，将前一层的输出从256维映射到128维（256 // 2）。同样使用Leaky ReLU激活函数。
3. **输出层（Output Layer）:**
   最后一层（fc4）: 一个全连接层，将前一层的输出从128维映射到1维。激活函数为Sigmoid，将输出值限制在[0, 1]之间，表示输入样本是真实数据的概率。

```python
class Discriminator(nn.Module):
    def __init__(self,input_size):
        super(Discriminator, self).__init__()
        self.fc1 = nn.Linear(input_size, 512)
        self.fc2 = nn.Linear(self.fc1.out_features, self.fc1.out_features//2)
        self.fc3 = nn.Linear(self.fc2.out_features, self.fc2.out_features//2)
        self.fc4 = nn.Linear(self.fc3.out_features, 1)
    def forward(self, x):
        x = F.leaky_relu(self.fc1(x), 0.2)
        x = F.leaky_relu(self.fc2(x), 0.2)
        x = F.leaky_relu(self.fc3(x), 0.2)
        x = torch.sigmoid(self.fc4(x))
        return x
D = Discriminator(input_size=28 * 28)
print("summary of D:")
torchsummary.summary(D,input_size=[(1,28*28)],batch_size=32,device="cpu")
```

### 五、GAN线性插值

生成两个不同的噪声向量，用一个噪声的值逐步替代另一个噪声，可以得到一系列的中间噪声向量，这些中间噪声向量对应的图像可以看做是原始图像的过渡状态，然后通过生成器G生成对应的图像，最终得到一系列中间图像。GAN 插值的目的是为了生成具有连续变化的图像序列，这样可以实现图像的平滑过渡。

```python
import torch
from matplotlib import pyplot as plt
from network import Generator, Discriminator

torch.manual_seed(1)
noise1 = torch.randn(1, 100)
noise2 = torch.randn(1, 100)

G = Generator(input_size=100, output_size=28 * 28)
D = Discriminator(input_size=28 * 28)
G.load_state_dict(torch.load("G_model.pt"))
D.load_state_dict(torch.load("D_model.pt"))

origin_img1 = G(noise1).view(1, 28, 28)
origin_img2 = G(noise2).view(1, 28, 28)
def linear_interpolate(v1, v2, t):
    if t == -1:
        return v1
    if t == 10:
        return v2
    img = v1.clone()
    img[:, 0:t * 10 + 10] = v2[:, 0:t * 10 + 10]
    return img
for i in range(-1,10):
    noise = linear_interpolate(noise1, noise2, i)
    with torch.no_grad():
        fake_img = G(noise)
        fake_img = fake_img.view(28, 28)
    plt.subplot(1, 11, i + 2)
    plt.imshow(fake_img, cmap="grey")
    plt.xticks([])
    plt.yticks([])
plt.show()
```

一个噪声初始图像为8，一个噪声初始图像为3，通过一系列替换产生连续过渡向量，然后经过生成器G生成对应的一系列中间图像，可以看到8逐步变换为3，不过中间过渡不是很平滑，加大训练轮数应该能改善过度不平滑的问题。

### 六、问题总结

GAN是由一个生成网络(generator)和判别网络(discriminator)组成，generator负责生成逼真数据以“骗”过discriminator，而discriminator负责判断一个样本是真实的还是“造”出来的。GAN的训练其实就是两者相互学习，相互博弈的过程，判别器因监督学习而提升，生成器因判别器而提升，同时两者都不能过于强大，导致另一方无法学习。

1. 判别器能力过强，使得D_loss基本为0而G_loss稳定在一个较大的值，生成器从中学习到的信息很少，导致训练网络梯度消失，无法学习。
2. 判别器能力过弱，使得D_loss稳定在一个较大的值，生成器无法稳定学习，丢失生成样本多样性，导致模型崩溃。

**解决方法：**

1. 软标签和带噪声的标签：在真实样本的标签1上加上(-0.2,0)的随机噪声，在虚假样本的标签上加(0, 0.2)的随机噪声，软标签和带噪声的标签的加入可以在一定程度上解决生成网络梯度消失的问题。
2. 反转标签：将部分真实标签和虚假标签反转，刺激网络的权重更新，跳过相对极值点，让网络能继续训练下去，打破相对极值点的僵局。
